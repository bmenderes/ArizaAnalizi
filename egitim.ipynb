```python
# prompt: oluuşturmuş olduğumuz bu rakamları sıraya sokabilir misin

import random

random_numbers = []
for _ in range(20):
  random_numbers.append(random.randint(1, 100)) # 1 ile 100 arasında rastgele sayılar

random_numbers.sort()
random_numbers
```

```python
# prompt: oluşturmuş olduğumuz bu rakamları bir tabloya döünştürebilir misin

import pandas as pd

# Assuming 'random_numbers' list is already defined from previous code

df = pd.DataFrame({'Random Numbers': random_numbers})
df
```

```python
# prompt: oluitumuş olduğumuz tabloyu histogram olarak bize grafikleştirir misin

import matplotlib.pyplot as plt

# Assuming 'df' DataFrame is already defined from previous code

plt.hist(df['Random Numbers'], bins=10, edgecolor='black')  # bins parametresi ile histogramın sütun sayısını ayarlayabilirsiniz
plt.xlabel('Sayılar')
plt.ylabel('Frekans')
plt.title('Rastgele Sayıların Histogramı')
plt.show()
```
